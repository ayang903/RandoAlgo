# Generate preferences, find perfect matches, calculate score, filter to stable matches

```python
# Generate preferences

# generate random integer 1-10, this is the number of people and partners
import random
import string

# n = random.randint(1, 7)

n=6 # I find when n=6 it is easy to get an example of non pareto optimal stable matches
lowercase_alphabet = list(string.ascii_lowercase)

people = lowercase_alphabet[:n] # "people" are the LOWERCASE letters
partners = [x.upper() for x in people] # "partners" are the UPPERCASE letters
```

```python
all_preferences = {}
for person in people: # loop through the lowercases
    partner_preference = random.sample(people, len(people))
    person_preference = random.sample(partners, len(partners)) # this is a preference for a lowercase guy, this list contains uppercase letters
    all_preferences[person] = person_preference
    all_preferences[person.upper()] = partner_preference
```

# Run this cell to manually set preferences to the one in the homework

```python
n=3
people = ['a', 'b', 'c']
partners = ['A', 'B', 'C']
all_preferences={
    'A': ['a', 'b', 'c'],
    'B': ['a', 'c', 'b'],
    'C': ['c', 'a', 'b'],
    'a': ['C', 'B', 'A'],
    'b': ['A', 'B', 'C'],
    'c': ['C', 'B', 'A']
    }
```

```python
# Generate all perfect matchings and calculate their scores

all_matches_and_scores = {}
initial_match = dict(zip(partners, people))

# swap two indices to keep generating perfect matches, add to all_matches_and_scores if never seen before

for _ in range(5_000_000): # arbitrarily picked 5 million

    # pick two people to swap
    idx1 = random.choice(partners)
    item1 = initial_match[idx1]
    idx2 = random.choice(partners)
    item2 = initial_match[idx2]

    # make the swap
    initial_match[idx1] = item2
    initial_match[idx2] = item1

    # This is just to convert the dictionary to a hashable type to store in the dict
    initial_match_tuple = tuple(sorted(initial_match.items()))

    # store and generate objective function score if it hasn't been seen before
    if initial_match_tuple not in all_matches_and_scores:
        total_point = 0
        for i in initial_match:
            print(f"matching {i} to {initial_match[i]}")
            points_cap = n - all_preferences[i].index(initial_match[i])
            print(f"capital letter point {points_cap}")
            points_low = n - all_preferences[initial_match[i]].index(i)
            print(f"lowercase letter point {points_low}")
            total_point += points_cap + points_low

        all_matches_and_scores[initial_match_tuple] = total_point
```

```python
import math

# n! possible perfect matches, make sure we generated all of them
assert(math.factorial(n) == len(all_matches_and_scores))
print(len(all_matches_and_scores))
```

```python
# Filter all perfect matchings down to only stable matchings

perfect_matchings = [dict(match) for match in all_matches_and_scores]

def is_stable(matching, preferences):
    for person, partner in matching.items():
        # print('---')

        # access the preferences for the person
        person_prefs = preferences[person]

        # print(f"Person {person} has preferences {person_prefs}")

        # find index of current partner
        partner_index = person_prefs.index(partner)

        # and check if there is someone else the person prefers more than the current partner
        for preferred_partner in person_prefs[:partner_index]:
            
            # Get find the preferred partner's current partner
            other_partner = {v: k for k, v in matching.items()}[preferred_partner]
            
            # print(f"Other partner: {other_partner}")

            # check if the preferred partner would also prefer the current person
            # "if the I appear higher on the preferred partner's list than their current partner", unstable
            if preferences[preferred_partner].index(person) < preferences[preferred_partner].index(other_partner):
                return False  
    return True


""" 
Essentially, we do this:

1. cycle thru every proposal
2. ask person, "hey, is there anyone at all that you like more than your current partner?"
3. yes? let's see if they like you more than their current parter as well...

case 1: hmm yeah so it seems like you are also higher on their list than their current partner, unstable
case 2: oops it looks like their current partner is higher on their list than u, they like them more than they like you. stable so far.
"""


stable_matchings = [matching for matching in perfect_matchings if is_stable(matching, all_preferences)]

len(stable_matchings), stable_matchings
```

```python
# get a list of the scores of these stable matchings

scores = []

for stablematch in stable_matchings:
    stablematch_key = tuple(sorted(stablematch.items()))
    score = all_matches_and_scores[stablematch_key]
    scores.append(score)
```
